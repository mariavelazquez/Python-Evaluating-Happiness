```python
import pandas as pd
import matplotlib.pyplot as plt

# Defining DataFrame
df = pd.read_csv('finalversionfile.csv')
```

```python
def runMenu(df):
    #Menu system
    quit = False
    while quit == False:
        print("\nMENU")
        print("1. List Column Types of DataFrame")
        print("2. Get Means of Quantitative Columns")
        print("3. Get Frequencies of Qualitative Columns")
        print("4. Generate Descriptive Statistics for Numeric Columns")
        print("5. Quit")
        menu_choice = input("What is your choice: ")
        try:
            menu_choice = int(menu_choice) #convert to integer
        except: #if the user enters text or a symbol
            print("ERROR: Please enter 1, 2, 3, 4 or 5")
            continue #returns to top of loop
        if menu_choice not in [1, 2, 3, 4, 5]:
            print("ERROR: Please enter 1, 2, 3, 4 or 5")
        else:
            if menu_choice == 1:
                print("\nList Column Types of DataFrame")
                print(df.dtypes)
            if menu_choice == 2:
                print("\nGet Means of Quantitative Columns")
                means = []
                columns = []
                for c in df.select_dtypes('number').columns:
                    columns.append(c)
                    mean = df[c].mean()
                    means.append(mean)
                    print(f"{c}: {mean:.2f}")
                plt.figure()
                plt.bar(columns, means)
                plt.xlabel("Columns")
                plt.ylabel("Mean")
                plt.title("Means of Quantitative Columns")
                plt.xticks(rotation=70)  # Rotate x-labels by 45 degrees
                plt.show()
            if menu_choice == 3:
                print("\nGet Frequencies of Qualitative Columns")
                freq = []
                columns = []
                for c in df.select_dtypes('object').columns:
                    if c == 'Country':
                        continue  # Skipping plotting the frequencies for the 'Country' column
                    print(f"{c}: \n\n{df[c].value_counts()}\n")
                    df[c].value_counts().plot(kind='bar')
                    plt.xlabel("Categories")
                    plt.ylabel("Frequency")
                    plt.title(f"Frequencies of {c}")
                    plt.show() 
            if menu_choice == 4:
                    print("\nGenerate Descriptive Statistics for Numeric Columns")
                    numeric_columns = df.select_dtypes('number').columns
                    descriptive_stats = df[numeric_columns].describe()
                    print(descriptive_stats)
            if menu_choice == 5:
                quit = True
                          
        
```

```python
#Option 1 + 5 (to quit)
runMenu(df)
```
